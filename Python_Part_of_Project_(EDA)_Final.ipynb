**NULL value Count and Percentage**

```python
null_counts = data.isnull().sum()
null_percentages = (data.isnull().sum() / len(data)) * 100

missing_data = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

display(missing_data.sort_values(by='Null Percentage', ascending=False))
```

**Deal with Null Values**

```python
# we will drop columns ['Product Description', 'Order Zipcode'] beacause it has high precentage of NULLs
data = data.drop(columns=['Product Description', 'Order Zipcode'])
```

```python
# we will drop columns ['Customer Password', 'Customer Email'] beacause it has no value (XXXXXXXX)
data = data.drop(columns=['Customer Password', 'Customer Email'])
```

```python
# we will drop columns ['Product Status'] beacause it all Values  = 0 , so it has no meaning
data = data.drop(columns=['Product Status'])
```

```python
# we will drop columns ['Customer Zipcode'] beacause it has no meaning
data = data.drop(columns=['Customer Zipcode'])
```

```python
#  The column Customer Lname has 8% NULL Valus ,so we will compensate with mode
lname_mode = data['Customer Lname'].mode()[0]
data['Customer Lname'] = data['Customer Lname'].fillna(lname_mode)
print(f"Customer Lname nulls remaining: {data['Customer Lname'].isna().sum()}")
```

**Standardise Customer Country encoding**

```python
data['Customer Country'] = data['Customer Country'].replace('EE. UU.', 'United States')
print("Customer Country values after fix:")
print(data['Customer Country'].value_counts())
```
